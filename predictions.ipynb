```python
import tensorflow as tf
from tensorflow.keras.models import load_model
import pickle
import pandas as pd
import numpy as np
```

```python
## Let's load trained model, scaler pickle file and onehot encoding
model=load_model('model.h5')

#Loader encoder and scaler
with open('label_encoder_gender.pkl','rb') as f:
    label_encoder_gender=pickle.load(f)

with open('ohe.pkl','rb') as f:
    ohegeo=pickle.load(f)

with open('scaler.pkl','rb') as f:
    scaler=pickle.load(f)
```

```python
input_data={
    'CreditScore':600,
    'Geography':'France',
    'Gender':'Male',
    'Age':40,
    'Tenure':3,
    'Balance':60000,
    'NumOfProducts':2,
    'HasCrCard':1,
    'IsActiveMember':1,
    'EstimatedSalary':50000
}
```

```python
geoencoded=ohegeo.transform([[input_data['Geography']]]).toarray()
geoencodeddf=pd.DataFrame(geoencoded,columns=ohegeo.get_feature_names_out(['Geography']))
```

```python
geoencodeddf

input_datadf=pd.DataFrame([input_data])
```

```python
input_data=pd.concat([input_datadf.reset_index(drop=True),geoencodeddf],axis=1)
```

```python
input_data
```

```python
input_data.drop(['Geography'],inplace=True,axis='columns')
```

```python
input_data['Gender']=label_encoder_gender.transform(input_data['Gender'])
input_data
```

```python
input_data
```

```python
##Now lets scale the input data
input_scaled=scaler.transform(input_data)
```

```python
input_scaled
```

```python
## Let's predict churn
prediction=model.predict(input_data)
```

```python
prediction
```

```python
if prediction>0.5:
    print('The customer is likely to churn')
else:
    print('Customer is not going to churn')
```
