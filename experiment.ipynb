```python
import pandas as pd 
from sklearn.model_selection import train_test_split
from sklearn.preprocessing  import StandardScaler,LabelEncoder
import pickle
```

```python
## Let's first load the dataset
data=pd.read_csv("D:\ProjectAarya\GEN AI\Code\Ch13\Churn_Modelling.csv")
data
```

```python
## It is clearly understandable that rownumber really does not play any role at all so we will delete that row number

data.drop(['RowNumber'],inplace=True,axis='columns')
```

```python

#Another thing is customerID surname also don't play any role at all so we will drop them too
data.drop(["CustomerId","Surname"],inplace=True,axis='columns')
```

```python
## Let's first encode the categorical variables
label_encoder_gender=LabelEncoder()
data['Gender']=label_encoder_gender.fit_transform(data['Gender'])
data
```

```python
import numpy as np
np.unique(data['Geography'])
```

```python
## Now we will work on 'Geography' column
from sklearn.preprocessing import OneHotEncoder
ohe=OneHotEncoder()
geoencoded=ohe.fit_transform(data[['Geography']])
geoencoded
```

```python
ohe.get_feature_names_out(['Geography'])
```

```python
geoencodeddense=geoencoded.toarray()
# geoencodeddf=pd.DataFrame(geoencoded,columns=ohe.get_feature_names_out(['Geography']))
geoencoded.shape
```

```python
geoencodeddf=pd.DataFrame(geoencodeddense,columns=ohe.get_feature_names_out())
```

```python
geoencodeddf
```

```python
data.drop(['Geography'],inplace=True,axis='columns')
```

```python
data=pd.concat([data,geoencodeddf],axis='columns')
```

```python
data
```

```python
## Save the encoder and scaler
with open('label_encoder_gender.pkl','wb') as file:
    pickle.dump(label_encoder_gender,file)

with open('ohe.pkl','wb') as file:
    pickle.dump(ohe,file)
```

```python
#Let's quickly divide the dataset into dependent and independent feature
X=data.drop('Exited',axis='columns')
y=data['Exited']

#Splitting the data in train test split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.14,random_state=6)
```

```python
##Scalling the features
scaler=StandardScaler()
X_train=scaler.fit_transform(X_train)
X_test=scaler.fit_transform(X_test)
```

### ANN Implementation

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.callbacks import EarlyStopping,TensorBoard
import datetime
```

```python
# ## Let's Build our Model
# model=Sequential([
#     Dense(32,activation='relu',input_shape=(X_train.shape[1],)),
#     Dense(16,activation='relu'),
#     Dense(1,activation='sigmoid')
# ])
# model.compile(
#     optimizer='adam',
#    loss='binary_crossentropy',
#     metrics=['accuracy']
# )
```

```python
# Use TensorFlow's Keras explicitly
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

model = Sequential([
    Dense(32, activation='relu', input_shape=(X_train.shape[1],)),
    Dense(16, activation='relu'),
    Dense(1, activation='sigmoid')
])

model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
model.summary()
```

```python
## setting up the tensorboard
from tensorflow.keras.callbacks import EarlyStopping,TensorBoard
log_dir="logs/fit"+datetime.datetime.now().strftime("%Y%m%d-%H%M%S")
tensorflow_callbacks=TensorBoard(log_dir=log_dir,histogram_freq=1)
```

```python
## Setting up early stopping
earlystop=EarlyStopping(monitor='val_accuracy',patience=9,restore_best_weights=True)
```

```python
## Let's train the actual model
history=model.fit(
    X_train,y_train,validation_data=(X_test,y_test),epochs=100,
    callbacks=[tensorflow_callbacks,earlystop]
)
```

```python
model.save('model.h5')
```
